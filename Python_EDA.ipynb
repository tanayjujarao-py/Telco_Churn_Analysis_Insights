```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("F:\Customer Churn.csv")
df.head()
```

```python
df.info()
```

```python
df["TotalCharges"] = df["TotalCharges"].replace(" ",0)
df["TotalCharges"] = df["TotalCharges"].astype("float")
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
df["customerID"].duplicated().sum()
```

```python
df["SeniorCitizen"] = df["SeniorCitizen"].replace(1,"yes")
df["SeniorCitizen"] = df["SeniorCitizen"].replace(0,"no")
df
```

```python
ax = sns.countplot(x = "Churn",data = df)
ax.bar_label(ax.containers[0])
plt.title('Count of Churned Customers',fontsize=13)
plt.show()
```

```python

gb = df.groupby("Churn").agg({"Churn":"count"})
plt.pie(gb["Churn"],labels = gb.index,autopct = "%1.2f%%")
plt.title("Perncentage of Churned Customers",fontsize=13)
plt.show()
```

```python
plt.figure(figsize=(4,4))
sns.countplot(x = 'gender',data = df,hue="Churn")
plt.title("Churn by Gender")
plt.show()
```

```python
plt.figure(figsize=(4,4))
bx = sns.countplot(x = 'SeniorCitizen',data = df)
bx.bar_label(bx.containers[0])
plt.title("Churn by Senior Citizen")
plt.show()
```

```python

counts = pd.crosstab(df['SeniorCitizen'], df['Churn'])

# Convert counts to percentages
percentages = counts.div(counts.sum(axis=1), axis=0) * 100

# Plot 100% stacked bar chart
ax = percentages.plot(
    kind='bar',
    stacked=True,
    figsize=(5, 5)
)

# Titles and labels
plt.title("Churn by Senior Citizen (%)")
plt.xlabel("Senior Citizen")
plt.ylabel("Percentage")
plt.legend(title="Churn")

# Add percentage labels inside bars
for container in ax.containers:
    ax.bar_label(
        container,
        fmt='%.1f%%',
        label_type='center'
    )

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(7,5))
sns.histplot(x = "tenure",data = df,hue="Churn")
plt.show()
```

```python
plt.figure(figsize=(5,5))
bx = sns.countplot(x = 'Contract',data = df,hue='Churn')
bx.bar_label(bx.containers[0])
bx.bar_label(bx.containers[1])
plt.title("Churn on Basis of Contract")
plt.show()
```

```python
cols = [
    'PhoneService', 'MultipleLines', 'InternetService',
    'OnlineSecurity', 'OnlineBackup', 'DeviceProtection',
    'TechSupport', 'StreamingTV', 'StreamingMovies'
]

# Create subplots (3 rows x 3 columns)
fig, axes = plt.subplots(3, 3, figsize=(15, 12))
axes = axes.flatten()

# Plot countplots
for ax, col in zip(axes, cols):
    sns.countplot(
        x=col,
        data=df,
        ax=ax,
        hue="Churn"
    )
    ax.set_title(col)
    ax.set_xlabel("")
    ax.set_ylabel("Count")
    ax.tick_params(axis='x', rotation=30)

# Remove empty subplots (safety)
for i in range(len(cols), len(axes)):
    fig.delaxes(axes[i])

plt.tight_layout()
plt.show()
```
